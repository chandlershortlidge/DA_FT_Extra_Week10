```python
import random

user_input = input("What type of music do you want to listen to? ")

if user_input.lower() == "trending":
    song = billboard_df.sample(1).iloc[0]
    query = f"{song['song_title']} {song['artist']}"
    
else:
    # Match user input to cluster label
    cluster_songs = clustered_df[clustered_df['cluster_label'].str.lower().str.contains(user_input.lower())]
    if len(cluster_songs) > 0:
        song = cluster_songs.sample(1).iloc[0]
        query = f"{song['track_name']} {song['artists']}"
    else:
        print("Category not found, try one of these:\n"
        "Upbeat Pop / Dance\n" "Soft Indie / Acoustic\n"  'International / World\n'
        'Mainstream Pop / Rock\n' 'Hard Rock / Anthemic\n' 'Alternative / Country Rock\n'
        'Classical / Instrumental\n' 'Techno / Dark Electronicn\n' 'Hip Hop / Reggaeton\n'
        'Latin Party / Reggaeton\n')
        query = None

if query:
    result = sp.search(q=query, limit=1, market="GB")
    track_id = result["tracks"]["items"][0]["id"]
    display(play_song(track_id))
```
